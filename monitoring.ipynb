```python
import os
import numpy as np
import matplotlib.pyplot as plt
os.chdir("../results")
```

```python
u_bulk = 0.051   ##### to change ######
```

```python
# ubulk
lx = 0.256

# load files
f = np.loadtxt('time_fx.dat')
u = np.loadtxt('time_ubulk.dat')

# define arrays
tf = f[:,0]
fx = f[:,1]
tu = u[:,0]
ub = u[:,1]

# plot fx
fav = np.mean(fx)
print("fx_average =", fav)
tend = np.max(tf)
print("t_end =", tend)

plt.figure(1)

#plt.subplot(211)
plt.plot(tf, fx, 'b-')
plt.axis([0, tend, fav-fav, fav+0.5*fav])

#plt.subplot(212)
#plt.plot(tu, ub, 'r-')
#plt.axis([0, 20, 0.0725, 0.08])

plt.show()

# channel perfusions
tx = lx/u_bulk
print("t_1 =", tx)
print("t_3 =", tx*3)
print("t_10 =", tx*10)
print("t_64 =", tx*64)

# average dt in time range t_ra of 10 channel perfusions

#t_ra = 350
t_ra = np.max(tf) - tx*10
t_ra1 = np.min(np.where(tf > t_ra))
t_ra2 = np.max(np.where(tf > t_ra))

dt = np.empty(t_ra2 - t_ra1)
j = 0
for i in range(t_ra1,t_ra2):
    dt[j] = tf[i + 1] - tf[i]
    j = j + 1
    
dt_av = np.mean(dt)
t_bin = int(lx/u_bulk*4/dt_av)

print("dt =", dt_av)
print("t_bin =", t_bin)
```
